Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}


#Проверка наличия номера документа в базе:
def check(number_doc):
    for document in documents:
        if document['number'] == number_doc:
            return "Find"
    return False 
   
import copy

# p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;
def name_by_number(number_doc):
    for document in documents:
        if document['number'] == number_doc:
            person = document['name']
            return person
    return('Данного документа нет в базе данных, проверьте корректность введенных данных.')

# s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится
def shelf_by_number(number_doc):
    for document in directories:
        if number_doc in directories[document]:
            return document # Вернет номер полки
    return('Данного документа нет в базе данных, проверьте корректность введенных данных.')

# l – list – команда, которая выведет список всех документов в формате: passport "2207 876234" "Василий Гупкин"
def list_of_doc():
    for document in documents:
        a = []
        for key in document.values():
            a.append(key)
        print(f'{a[0]} "{a[1]}" "{a[2]}"')

# as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень.
def add_shelf(number_shelf):
    list_shelfs = []
    for key in directories.keys():
        list_shelfs.append(key)
    if number_shelf in list_shelfs:
        return False
    else: 
        directories[number_shelf] = []
        list_shelfs.append(number_shelf)
        print(f"Полка '{number_shelf}' добавлена. Список всех доступных полок: {list_shelfs}") 
        
# d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок.        
def delete_doc(number_doc):
    if check(number_doc) == False:
        print(f'Документа с номером {number_doc} нет в базе данных. Проверьте корректность ввода.')
    else:
        for document in documents:
              if document['number'] == number_doc:
                    del(documents[documents.index(document)])
        var = directories[shelf_by_number(number_doc)] 
        shelf_ = copy.copy([shelf_by_number(number_doc)])
        for i, docum_ in enumerate(var):
            del(var[i-1])
        directories[shelf_by_number(number_doc)] = var
        print(f'Документ с номером {number_doc} удален c полки {shelf_}, на ней остались документы: {directories[shelf_by_number(number_doc)]}')

# m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую.
def move_doc(number_doc, number_shelf):
    if check(number_doc) == False:
        print(f'Документа с номером {number_doc} нет в базе данных. Проверьте корректность ввода.')
    else:
        # Создаю var, чтобы сначала использовать его для нахождения списка книг на той, полке, где лежала целевая книга. Потом
        # удаляю эту книгу со старой полки (через индекс var[i-1]) и перезаписываю список книг на новую по ключу = номер полки.
        var = directories[shelf_by_number(number_doc)] 
        shelf_ = copy.copy(shelf_by_number(number_doc))
        for i, docum_ in enumerate(var):
            del(var[i-1])
        directories[shelf_] = var
        #Создаем список всех полок в а, а потом проверяем существует ли сейчас полка, на которую нужно переложить, 
        # если не существует- выполняем команду с созданием новой полки, передавая в него номер, который указал пользователь
        a = []
        for key in directories.keys():
                a.append(key)
        if number_shelf not in a:
            add_shelf(number_shelf)
        # Добавляем на целевую полку книжку, путем переприсваивания значения по ключу.
        var = directories[number_shelf] 
        var.append(number_doc)
        directories[number_shelf] = var
        print(f'Документ с номером "{number_doc}" перемещен на полку: {number_shelf}. Ее содержимое:{directories[number_shelf]}')       

# a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив: 
# его номер, тип, имя владельца и номер полки, на котором он будет храниться.
def add_doc_to_catalog(number_doc, type_, name, number_shelf):
    if check(number_doc) != False:
        print(f'Документа с номером {number_doc} уже существует в базе.')
    else:
        add_dict = dict([('type', type_), ('number', number_doc), ('name', name)])
        documents.append(add_dict)
        if add_shelf(number_shelf) == False:
            var = directories[number_shelf]
            var.append(number_doc)
            directories[number_shelf] = var
            print(f'Документ {add_dict} добавлен на полку: {number_shelf}.')
        else:
            add_shelf(number_shelf)
            var = directories[number_shelf]
            var.append(number_doc)
            directories[number_shelf] = var
            print(f'Документ {add_dict} добавлен на полку: {number_shelf}.') 
            
def main():
    def choice():
        Next = input("Хотите ли Вы продолжить работу с документами? 'Да'/'Нет'")
        no = ["нет", "Нет", "no", "No", "ytn", "Ytn"]
        if Next  in no:
            print(f'Работа завершена. Пользоватеьские команды выполнены')
        else:
                main()
    comand_number = input('Введите номер команды, которую необходимо осуществить: p. Вывести имя человека по номеру паспорта s. Вывести номер полки, на которой находится документ l. Показать список всех документов. as. Добавить новую полку. d. Удалить документ m. Переместить документ add. Добавить документ. ')
    if comand_number == "p":
        number_doc = input('Введите номер документа: ')
        print(name_by_number(number_doc))
        choice()
    elif comand_number == "s":
        number_doc = input('Введите номер документа: ')
        print(f' Документ находится на полке: {shelf_by_number(number_doc)}')
        choice()
    elif comand_number == "l":
        print(f'Список всех документов:')
        list_of_doc()
        choice()
    elif comand_number == "as":
        number_shelf = input('Введите номер полки: ')    
        if add_shelf(number_shelf) == False:
            print(f'Данный номер полки уже существует, выберите другую полку.')
        add_shelf(number_shelf)
        choice()
    elif comand_number == "d":
        number_doc = input('Введите номер документа: ')
        delete_doc(number_doc)
        choice()
    elif comand_number == "m":
        number_doc = input('Введите номер документа: ')
        number_shelf = input('Введите номер полки: ') 
        move_doc(number_doc, number_shelf)
        choice()
    elif comand_number == "add":
        number_doc = input('Введите номер документа: ')
        if check(number_doc) != False:
            print(f'Документа с номером {number_doc} уже существует в базе.')
            choice()
        else:
            type_ = input('Введите тип: ')
            name = input('Введите имя: ')
            number_shelf = input('Введите номер полки: ')
            add_doc_to_catalog(number_doc, type_, name, number_shelf)
        choice()
main()
```
